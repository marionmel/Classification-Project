#### Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
file_path = "churn.xlsx"
df = pd.read_excel(file_path, sheet_name="Telco_Churn")
#load dataset
```

```python
df["Total Charges"] = pd.to_numeric(df["Total Charges"], errors="coerce")
#convert 'total charges' to numeric (some might be stored as strings)
```

```python
missing_values = df.isnull().sum()
print(missing_values)
#check missing values
```

```python
numerical_summary = df.describe()
print(numerical_summary)
#summary statistics for numerical features
```

```python
churn_counts = df["Churn Label"].value_counts()
#churn distribution
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(x=churn_counts.index, y=churn_counts.values)
plt.xlabel("Churn Label")
plt.ylabel("Count")
plt.title("Churn Distribution")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
#correlation matrix
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df["Churn Label"], y=df["Monthly Charges"])
plt.xlabel("Churn Label")
plt.ylabel("Monthly Charges")
plt.title("Monthly Charges Distribution by Churn")
plt.show()
#boxplot of monthly charges by churn
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df["Churn Label"], y=df["Tenure Months"])
plt.xlabel("Churn Label")
plt.ylabel("Tenure Months")
plt.title("Tenure Months Distribution by Churn")
plt.show()
#boxplot of tenure months by churn
```
